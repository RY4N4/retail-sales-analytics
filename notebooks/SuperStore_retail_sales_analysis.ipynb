STEP 3: ANALYTICS MODELING WITH SQL (STAR SCHEMA)

IMPORT LIBRARIES

```python
import pandas as pd
import sqlite3
```

LOAD FEATURE-ENGINEERED DATA (FROM STEP 2)

```python
df = pd.read_csv("data_cleaned/cleaned_sales.csv")
df.head()
```

CREATE SQLITE DATABASE

```python
conn = sqlite3.connect("retail_analytics.db")
cursor = conn.cursor()
```

LOAD RAW DATA INTO A STAGING TABLE


```python
df.to_sql("stg_sales", conn, if_exists="replace", index=False)
```

```python
# verify
pd.read_sql("SELECT COUNT(*) AS row_count FROM stg_sales", conn)
```

CREATE DIMENSION TABLES

```python
cursor.execute("""
CREATE TABLE dim_date AS
SELECT DISTINCT
    "Order Date"        AS order_date,
    "Order Year"        AS year,
    "Order Quarter"     AS quarter,
    "Order Month"       AS month,
    "Order Month Name"  AS month_name,
    "Order YearMonth"   AS year_month
FROM stg_sales;

""")
conn.commit()
```

```python
cursor.execute("""
CREATE TABLE dim_product AS
SELECT DISTINCT
    "Product ID"       AS product_id,
    "Product Name"     AS product_name,
    Category            AS category,
    "Sub-Category"     AS sub_category
FROM stg_sales
""")
conn.commit()
```

```python
cursor.execute("""
CREATE TABLE dim_customer AS
SELECT DISTINCT
    "Customer ID"      AS customer_id,
    "Customer Name"    AS customer_name,
    Segment             AS segment
FROM stg_sales
""")
conn.commit()
```

```python
cursor.execute("""
CREATE TABLE dim_region AS
SELECT DISTINCT
    Region                AS region,
    State                 AS state,
    City                  AS city,
    "Postal Code"         AS postal_code
FROM stg_sales
""")
conn.commit()
```

CREATE FACT TABLE

```python
cursor.execute("""
CREATE TABLE fact_sales AS
SELECT
    "Order ID"       AS order_id,
    "Order Date"     AS order_date,
    "Product ID"     AS product_id,
    "Customer ID"    AS customer_id,
    Region           AS region,
    Sales            AS sales,
    Revenue          AS revenue,
    Quantity         AS quantity,
    Discount         AS discount,
    Profit           AS profit,
    CASE
        WHEN Sales != 0 THEN Profit * 1.0 / Sales
        ELSE 0
    END              AS profit_margin
FROM stg_sales
""")
conn.commit()
```

VALIDATE STAR SCHEMA

```python
for table in ["fact_sales", "dim_date", "dim_product", "dim_customer", "dim_region"]:
    print(table)
    print(pd.read_sql(f"SELECT COUNT(*) AS cnt FROM {table}", conn))
    print("-" * 30)
```

CREATE KPI VIEWS

```python
cursor.execute("""
CREATE VIEW IF NOT EXISTS vw_monthly_revenue AS
SELECT
    d.year,
    d.month,
    d.month_name,
    SUM(f.revenue) AS total_revenue
FROM fact_sales f
JOIN dim_date d
    ON f.order_date = d.order_date
GROUP BY d.year, d.month, d.month_name
""")
conn.commit()
```

```python
df_monthly_revenue = pd.read_sql(
    "SELECT * FROM vw_monthly_revenue",
    conn
)
```

```python
cursor.execute("""
CREATE VIEW IF NOT EXISTS vw_region_performance AS
SELECT
    r.region,
    SUM(f.revenue) AS revenue,
    SUM(f.profit) AS profit,
    AVG(f.profit_margin) AS avg_margin
FROM fact_sales f
JOIN dim_region r
    ON f.region = r.region
GROUP BY r.region
""")
conn.commit()
```

```python
df_region_performance = pd.read_sql(
    "SELECT * FROM vw_region_performance",
    conn
)
```

```python
cursor.execute("""
CREATE VIEW IF NOT EXISTS vw_product_contribution AS
SELECT
    p.product_name,
    SUM(f.revenue) AS revenue,
    SUM(f.revenue) * 1.0 /
        (SELECT SUM(revenue) FROM fact_sales) AS revenue_share
FROM fact_sales f
JOIN dim_product p
    ON f.product_id = p.product_id
GROUP BY p.product_name
""")
conn.commit()
```

```python
df_product_contribution = pd.read_sql(
    "SELECT * FROM vw_product_contribution",
    conn
)
```

```python
# SELECT name, type FROM sqlite_master WHERE type='view';
pd.read_sql("SELECT * FROM vw_monthly_revenue ORDER BY total_revenue DESC LIMIT 10", conn)
```

FORECASTING


```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
# LOAD CLEANED DATA
df = pd.read_csv("data_cleaned/cleaned_sales.csv")
df.head()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
# AGGREGATE TO MONTHLY DEMAND
monthly_sales = (
    df
    .groupby(pd.Grouper(key="Order Date", freq="M"))
    .agg({
        "Revenue": "sum",
        "Quantity": "sum"
    })
    .reset_index()
)
```

```python
monthly_sales.head()
```

```python
# CREATE TIME INDEX
monthly_sales["time_index"] = np.arange(len(monthly_sales))
```

```python
# TRAINTEST SPLIT - Last 6 months = future we want to predict
train = monthly_sales.iloc[:-6]
test = monthly_sales.iloc[-6:]
```

```python
X_train = train[["time_index"]]
y_train = train["Revenue"]

X_test = test[["time_index"]]
y_test = test["Revenue"]
```

```python
# TRAIN BASELINE FORECAST MODEL
model = LinearRegression()
model.fit(X_train, y_train)

# EVALUATE MODEL
test_predictions = model.predict(X_test)

mae = mean_absolute_error(y_test, test_predictions)
mae
```

```python
# FORECAST NEXT 6 MONTHS
future_index = np.arange(
    monthly_sales["time_index"].max() + 1,
    monthly_sales["time_index"].max() + 7
).reshape(-1, 1)

future_predictions = model.predict(future_index)
```

```python
# Create and Store into a DataFrame
future_dates = pd.date_range(
    start=monthly_sales["Order Date"].max() + pd.offsets.MonthEnd(1),
    periods=6,
    freq="M"
)

forecast_df = pd.DataFrame({
    "Order Date": future_dates,
    "Forecasted Revenue": future_predictions
})
```

```python
# VISUALIZE ACTUAL VS FORECAST
plt.figure(figsize=(12, 6))

plt.plot(monthly_sales["Order Date"], monthly_sales["Revenue"], label="Actual Revenue")
plt.plot(forecast_df["Order Date"], forecast_df["Forecasted Revenue"],
         linestyle="--", marker="o", label="Forecasted Revenue")

plt.title("Monthly Revenue Forecast")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Save the the results
forecast_df.to_csv("data_cleaned/revenue_forecast.csv", index=False)
```
